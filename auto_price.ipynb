# Making Simple Linear Regression Model

```python
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
```

```python
x = df.loc[:,df.columns == 'enginesize']
y = df.loc[:,df.columns == 'price']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=0) 
```

```python
len(x_train), len(y_train), len(x_test), len(y_test) 
```

```python
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
r2_score=model.score(x_train,y_train)
r2_score
```

```python
model.intercept_ #intercept
```

```python
model.coef_ #slope
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn import metrics
```

```python
rmse=np.sqrt(metrics.mean_squared_error(y_test,y_pred))#RMSE
rmse
```

```python
 from sklearn.metrics import r2_score
```

```python
r2 = r2_score(y_test, y_pred)*100  #r2_score
r2
```

```python
print("The RMSE for the model is {} and r2 is {}%".format(round(rmse,2),round(r2,2)))
```

# Making Multiple Linear Regression Model 

```python
I = df.loc[:,['enginesize','stroke','compressionratio','peakrpm']]
T = df.loc[:,['price']]
```

```python
lr = LinearRegression()
```

```python
I_train, I_test, T_train, T_test = train_test_split(I, T, test_size=0.2, random_state=0) 
```

```python
len(I_train), len(T_train), len(I_test), len(T_test) 
```

```python
lr.fit(I_train, T_train)
```

```python
T_pred = lr.predict(I_test)
```

```python
rmse_ = np.sqrt(metrics.mean_squared_error(T_test,T_pred))#RMSE
rmse_
```

```python
r2_= r2_score(T_test, T_pred)*100  #r2_score
r2_
```

```python
print("The RMSE for the model is {} and r2 is {}%".format(round(rmse_,2),round(r2_,2)))
```

```python
#The R2 score is increased so this is our final model with accuracy of 83.63%.
```
